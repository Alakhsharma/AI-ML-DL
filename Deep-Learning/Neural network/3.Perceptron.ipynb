```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from prettytable import PrettyTable

def yin(x,w):
    z = np.dot(x,w)
    return z

 
def threshold(z):
    if (z>0):
        return 1
    elif (z==0):
        return 0
    else:
        return -1;

def fit(X_train, target, lr):
    # weight = np.array([1, 0, -1, -0.5])
    n, dim = X_train.shape
    
    weight = np.zeros(dim)
    bias = 0
    
    flag = 1  
    
    while(flag!=0): 
        t = PrettyTable(['x1', 'x2', '1', 'w1', 'w2', 'b', 'error'])

        flag=0
        for j in range(n):
            z = yin(X_train[j,:], weight)+ bias
            out = threshold(z)
            error = target[j]-out
            
            for k in range(dim):
                weight[k] = weight[k] + lr*(error)*X_train[j,k]
            bias = bias + lr*(error)
            t.add_row([X_train[j, 0], X_train[j, 1], '1', weight[0], weight[1], bias, error])
            if(error!=0):
                flag=1
        print(t)
       
    return [weight, bias]


def predict(X_test,y_test, weight, bias):
    ctr = 0
    for i in range(len(X_test)):
        z = yin(X_test[i,:], weight) + bias
        out = threshold(z)
        if (out == y_test[i]):
            ctr+=1
    
    print('Accuracy = '+ str(ctr/len(X_test)))

if __name__ ==  "__main__":
    X_train = np.array([[1,1],
                      [1,-1],
                      [-1,1],
                      [-1,-1]])
    
    y_train = np.array([-1,1,-1,-1])
    
    lr = 1
    
    weight, bias = fit(X_train, y_train, lr)
    
    predict(X_train,y_train,weight, bias)

    
    
    
    
```
