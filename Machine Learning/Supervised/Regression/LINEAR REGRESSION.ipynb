```python
data= pd.read_csv("C:/Users/chamo/Documents/Linear Regression.csv")
```

```python
X = data['X'].values
Y = data['Y'].values



n = len(X)

sumx= 0
sumx2= 0
sumy= 0
sumxy=0

for i in range(n):
    sumx= sumx+X[i]
    sumy= sumy+Y[i]
    sumx2= sumx2+(X[i]*X[i])
    sumxy= sumxy+(X[i]* Y[i])

b= (n * sumxy - (sumx * sumy))/((n*sumx2) - (sumx * sumx))
a= ((sumy* sumx2)-(sumx*sumxy))/((n*sumx2) - (sumx * sumx))
    
# a is intercept
# b is slope or regression coefficient
#printing the coefficient
print(a,b)
```

```python
print(f" f(x) = {b} X + {a}")
```
